```python
stateRollouts = {}
```

```python
group1States = pickle.load(open('group1States', 'rb'))
group1States = reconstructStates(group1States)
for state in group1States:
    rollouts, values = getActions(state, 1, 200000, allRollouts = True)
    for i in range(len(rollouts)):
        stateRollouts[(state, rollouts[i][0])] = (rollouts[i], values[i])
```

```python
group3States = pickle.load(open('group3States', 'rb'))
group3States = reconstructStates(group3States)
for state in group3States:
    rollouts, values = getActions(state, 1, 200000, allRollouts = True)
    for i in range(len(rollouts)):
        stateRollouts[(state, rollouts[i][0])] = (rollouts[i], values[i])
```

```python
generalTestStates = pickle.load(open('general Test States', 'rb'))
generalTestStates = reconstructStates(generalTestStates)
for state in generalTestStates:
    rollouts, values = getActions(state, 1, 200000, allRollouts = True)
    for i in range(len(rollouts)):
        stateRollouts[(state, rollouts[i][0])] = (rollouts[i], values[i])
```

```python
bestRollout = {}
for state, action in stateRollouts.keys():
    if state not in bestRollout:
        bestRollout[state] = stateRollouts[(state, action)]
    else:
        if stateRollouts[(state, action)][1] > bestRollout[state][1]:
            bestRollout[state] = stateRollouts[(state, action)]
```

```python
datas = []
def process(name, expgroup):
    d = {}
    f = open("playdata/" + name, "rb")
    f2 = open("playdata/" + name + "2", "rb")
    data = pickle.load(f)
    data2 = pickle.load(f2)
    d['name'] = name
    d['experiment group'] = expgroup
    d['mistake type'] = data['FatalFlawData'][-1]
    d['group'] = data['group']
    testStates = reconstructStates(pickle.load(open("group" + str(data['group'])+ "States", "rb")))
    total = 0
    for i in range(len(testStates)):
        for j in range(len(testStates)):
            if testStates[i] == data['testStates'][j][0]:
                action = data['testStates'][j][1]
                hRollout = stateRollouts[(testStates[i], action)][0]
                rRollout = bestRollout[testStates[i]][0]
                if hRollout[0] == rRollout[0]:
                    d[i + 1] = None
                else:
                    d[i + 1] = categorize(testStates[i], action, rRollout, hRollout)
    testStates = reconstructStates(pickle.load(open("General Test States", "rb")))
    for i in range(6, 12):
        for j in range(6):
            if testStates[i - 6] == data2['testStates'][j][0]:
                action = data2['testStates'][j][1]
                hRollout = stateRollouts[(testStates[i - 6], action)][0]
                rRollout = bestRollout[testStates[i - 6]][0]
                if hRollout[0] == rRollout[0]:
                    d[i + 1] = None
                else:
                    d[i + 1] = categorize(testStates[i - 6], action, rRollout, hRollout)
    datas.append(d)

    
    

process("calvin", "short explanation")
process("raymond", "short explanation")
process("kenneth", "short explanation")
process("sam", "short explanation")
process("stanley", "short explanation")
process("arvin", "short explanation")
process("hong", "short explanation")



process("aayush", "long explanation")
process("karthik", "long explanation")
process("nadir", "long explanation")
process("rustie", "long explanation")
process("andy", "long explanation")
process("gokul", "long explanation")


process("jonny", "no explanation")
process("alex", "no explanation")
process("ket", "no explanation")
process("charlie", "no explanation")
process("aakash", "no explanation")
process("neel", "no explanation")
process("kartik", "no explanation")
process("sera", "no explanation")
process("jinsu", "no explanation")


df = pandas.DataFrame(datas)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 
```
